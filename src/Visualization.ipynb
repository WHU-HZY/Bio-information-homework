```python
import umap.umap_ as umap
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
# UMAP 可视化
run = 1
umap = umap.UMAP(n_components=2, random_state=0)
X = pd.read_csv(f"/Users/huzhangyi/Downloads/scMDC-1.0.1/script/atac_pbmc10k/{run}_embedding.csv", sep=',', header=None).values
y_pred = pd.read_csv(f"/Users/huzhangyi/Downloads/scMDC-1.0.1/script/atac_pbmc10k/{run}_pred.csv", sep=',', header=None).values.flatten().astype(int)
X_umap2 = umap.fit_transform(X)
# Plotting
fig, axe = plt.subplots(1, 1, figsize=(10, 10))
# UMAP plot
sns.scatterplot(x=X_umap2[:, 0], y=X_umap2[:, 1], hue=y_pred, palette='tab10', ax=axe)
axe.set_title('Multiple Integretion Autoencoder')
axe.legend(loc=1)
plt.tight_layout()
plt.show()
```
